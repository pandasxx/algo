```python
#root_dir = "/home/sakulaki/1216_origin/tct_data1216/"
#selected_class = ["ASCUS", "LSIL", "ASCH", "HSIL", "SCC"]
root_dir = "/home/super-speed-data/train-data-yolov3/20181016/train"
selected_class = ["ASCUS", "LSIL", "ASCH", "HSIL", "SCC",
                  "AGC", "EC", "FUNGI", "TRI", "CC",
                  "ACTINO", "VIRUS"]

each_classes_num = [0 for i in range(len(selected_class))]

xml_file_dirs = os.listdir(root_dir)
xml_file_list = []
for xml_file_dir in xml_file_dirs:
    temp_xml_file_list = os.listdir(root_dir + xml_file_dir)
    for i, xml_file in enumerate(temp_xml_file_list):
        temp_xml_file_list[i] = root_dir + xml_file_dir + "/" + xml_file
    xml_file_list += temp_xml_file_list
    
for i, xml_file in enumerate(xml_file_list):
    xml_file_list[i] = xml_file.split(".")[0]

xml_file_list = list(set(xml_file_list))

for i, xml_file in enumerate(xml_file_list):
    xml_file_list[i] = xml_file + ".xml"
```

```python
def parse_xml_gen_data(xml_name, box_data, selected_classes):
    DOMTree = xml.dom.minidom.parse(xml_name)
    collection = DOMTree.documentElement
    filename = collection.getElementsByTagName("filename")
    objects = collection.getElementsByTagName("object")
    
    w = collection.getElementsByTagName("width")[0]
    w_val = int(w.firstChild.nodeValue)
    h = collection.getElementsByTagName("height")[0]
    h_val = int(h.firstChild.nodeValue)
    w.firstChild.replaceWholeText(str(h_val))
    h.firstChild.replaceWholeText(str(w_val))
    
    for object in objects:
        if object.getElementsByTagName("name")[0].firstChild.nodeValue in selected_classes:
            xmin = object.getElementsByTagName("xmin")[0]
            xmin_val = int(xmin.firstChild.nodeValue)
            xmax = object.getElementsByTagName("xmax")[0]
            xmax_val = int(xmax.firstChild.nodeValue)
            ymin = object.getElementsByTagName("ymin")[0]
            ymin_val = int(ymin.firstChild.nodeValue)
            ymax = object.getElementsByTagName("ymax")[0]
            ymax_val = int(ymax.firstChild.nodeValue)
        
            box_h = ymax_val - ymin_val
            box_w = xmax_val - xmin_val
        
            box_data.append([box_w, box_h])

def statistics_cell_num(xml_name, each_classes_num, selected_classes):
    DOMTree = xml.dom.minidom.parse(xml_name)
    collection = DOMTree.documentElement
    filename = collection.getElementsByTagName("filename")
    objects = collection.getElementsByTagName("object")
    
    for object in objects:
        name = object.getElementsByTagName("name")[0].firstChild.nodeValue
        if object.getElementsByTagName("name")[0].firstChild.nodeValue in selected_classes:
            index = selected_classes.index(name)
            each_classes_num[index] = each_classes_num[index] + 1
```

```python
box_data = []
count = 0

each_classes_num = [0 for i in range(len(selected_class))]

for i in range(int(len(xml_file_list)/1000)):
    for xml_file_path in xml_file_list[i * 1000:(i + 1) * 1000]:
        #parse_xml_gen_data(xml_file_path, box_data, selected_class)
        statistics_cell_num(xml_file_path, each_classes_num, selected_class)
    count += 1
    print("du", count)
    if (i == (int(len(xml_file_list)/1000) - 1)):
        for xml_file_path in xml_file_list[(i + 1) * 1000:]:
            parse_xml_gen_data(xml_file_path, box_data, selected_class)
```
